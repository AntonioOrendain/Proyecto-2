```python
x_recta = np.linspace(0, 2.88, 100)
y_recta = x_recta*1.73611 #m


x_radio=np.array([0,0])#Datos de radio
y_radio=np.array([0,7.3])

x=np.array([2.88,6.92]) #puntos dados
y=np.array([5,5])



x_circulo = np.linspace(0,6.92,100)
y_circulo = (-np.sqrt(7.3**2-x_circulo**2))+(7.3)


plt.figure(figsize = (12, 8))
plt.plot(x_recta,y_recta,'C1')
plt.plot(x_radio,y_radio,"C6",ms=20,label="Radio =(7.3)")
plt.plot(x_radio,y_radio,"C6")
plt.plot(x,y,"C1")
plt.plot(x,y,"y*",ms=20,label="Punto 1=(2.88,5),Punto2=(6.92,5)")
plt.plot(x_circulo,y_circulo,'C1')
plt.legend(loc="best")
plt.grid()
plt.show()
```

```python
def montecarlo2(region,a1,b1,a2,b2,N):
    AR = (b1-a1)*(b2-a2)
    x = np.random.uniform(a1,b1,N)
    y = np.random.uniform(a2,b2,N)
    b = region(x,y)
    return AR*np.mean(b)  
```

```python
N= 100
x=np.linspace(0,8,N)
y=np.linspace(0,5.5,N)
```

```python
def R1(x,y):
    return ((0<x) & (y<5) & (y<x*1.73611) & ((x**2)+((y-7.3)**2)<(7.3**2)))
```

```python
X , Y= np.meshgrid(x,y)
plt.figure(figsize=(5,5))
plt.scatter(X,Y,c=~R1(X,Y),cmap='bone')
plt.show()
```

```python
a1,a2,b1,b2=0,0,8,5.5
N=np.logspace(1,6,6).astype(int)

df= pd.DataFrame(index=N,columns=['Aprox'])

for n in N:
    df.loc[n,'Aprox'] = montecarlo2(R1,a1,b1,a2,b2,n)
df
```
